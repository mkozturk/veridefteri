Yaklaşık 2.200 yıldır denklem sistemlerini çözmeye çalışıyoruz
[1]. Hemen öyle karmaşık sistemler gelmesin aklınıza. Doğrusal
eşitliklerden oluşan, görece kolay sistemlerin çözülmeleri bile hâlâ
büyük bir olay. Çünkü mühendislik ve doğa bilimleri başta olmak üzere,
hesaplamalı işlerin tamamında daima karşımıza bir denklem sistemi
çıkıyor. Sadece her geçen yıl daha büyük sistemleri çözüyoruz. Fakat
ne kadar devasa olurlarsa olsunlar, neticede bu sistemler dönüp
dolaşıp bizim oğlanın eve getirdiği şu ödeve benziyor:

"Ayşe, annesi ve babasının yaşları toplamı 80'dir. Ayşe'nin ve
 babasının yaşlarının 2 katları ile annesinin yaşını toplayınca 120
 ediyor. Ayşe'nin yaşının 3 katı ile babasının yaşının 2 katının
 toplamından, annesinin yaşını çıkarınca geriye 48 kalıyor. Ayşe'nin,
 annesinin ve babasının yaşları kaçtır?"

Hah, çok kolay. Matematik Dünyası okuyucusu basit cebirden mi
korkacak? Ayşe'nin, babasının ve annesinin yaşları sırasıyla, $x_1$,
$x_2$ ve $x_3$ olsun. Altı üstü üç bilinmeyenli üç denklemi çözeceğiz:

$$
\begin{array}{rcl}
  x_1 + x_2 + x_3 & = & 80, \\
  2x_1 + 2x_2 + x_3 & = & 120, \\
  3x_1 + 2x_2 - x_3 & = & 48.
\end{array}
$$

Matris cebiriyle ifade edersek

$$
  \underset{A}{\underbrace{\left[\begin{array}{rrr}
    1 & 1 & 1 \\
    2 & 2 & 1 \\
    3 & 2 & -1
  \end{array}\right]}}\underset{x}{\underbrace{\left[
            \begin{array}{r}
              x_1 \\
              x_2 \\
             x_3
            \end{array}
            \right]}}=\underset{d}{\underbrace{\left[
            \begin{array}{c}
              80 \\
              120 \\
              48
            \end{array}
            \right]}}.
$$

Yani

$$
  Ax = d \hspace{10mm} (1)
$$

ile gösterilen ve çözülmeyi bekleyen bir sistem var karşımızda. Ya
sonra? Liseden hatırlarsınız belki. Satır işlemleri ile
bilinmeyenlerden birini çekmek için çeşitli numaralar
yapılabilir. Biraz uzun iş. Oysa birisi bize $A$ matrisini şöyle
çarpanlarına ayırıp verse ne güzel olur:

$$
  \underset{A}{\underbrace{\left[\begin{array}{rrr}
                                   1 & 1 & 1 \\
                                   2 & 2 & 1 \\
                                   3 & 2 & -1 
  \end{array}\right]}} = \underset{B}{\underbrace{\left[\begin{array}{rrr}
    \tfrac{1}{3} & \tfrac{1}{2} & 1 \\
    \tfrac{2}{3} & 1 & 0 \\
    1 & 0 & 0
  \end{array}\right]}} \underset{C}{\underbrace{\left[\begin{array}{rrr}
    3 & 2 & -1 \\
    0 & \tfrac{2}{3} & \tfrac{5}{3} \\
    0 & 0 & \tfrac{1}{2}
  \end{array}\right].}}
$$

Güzel olur çünkü hayatımızı çok kolaylaştırır. O sayede çözeceğimiz
(1) sistemini

$$
Ax = B\underset{y}{\underbrace{Cx}} = d
$$

şeklinde baştan yazabiliriz. Bir kenara $Cx = y$ diye not alıp,
$By = b$ sistemine bir bakalım:

$$
    \underset{B}{\underbrace{\left[\begin{array}{rrr}
    \tfrac{1}{3} & \tfrac{1}{2} & 1 \\
    \tfrac{2}{3} & 1 & 0 \\
    1 & 0 & 0
  \end{array}\right]}}  \underset{y}{\underbrace{\left[
            \begin{array}{r}
              y_1 \\
              y_2 \\
             y_3
            \end{array}
            \right]}} =   \underset{d}{\underbrace{\left[
            \begin{array}{c}
              80 \\
              120 \\
              48
            \end{array}
            \right]}}.
$$

$B$ matrisinin köşegeni altındaki değerler 0 olduğu için sırayla
değerleri yerine koyarak sistemi çözebiliriz. En son satırdan
başlarsak

$$
\begin{array}{l}
  y_1 = 48,\\
  \tfrac{2}{3}y_1 + y_2 = 120 \implies y_2 = 120 - 32  = 88, \\
  \tfrac{1}{3}y_1 + \tfrac{1}{2}y_2 + y_3= 80 \implies y_3 = 80 -
    16 - 44 = 20,
\end{array}
$$

işlemleriyle $y$ vektörünü bulmuş oluruz. Bir kenara $Cx=y$
yazmıştık. O zaman

$$
  \underset{C}{\underbrace{\left[\begin{array}{rrr}
        3 & 2 & -1 \\
        0 & \tfrac{2}{3} & \tfrac{5}{3} \\
        0 & 0 & \tfrac{1}{2}
      \end{array}\right]}}  \underset{x}{\underbrace{\left[
            \begin{array}{r}
              x_1 \\
              x_2 \\
             x_3
            \end{array}
            \right]}} =  \underset{y}{\underbrace{\left[
            \begin{array}{r}
              48 \\
              88\\
             20
            \end{array}
            \right]}}
$$

eşitliğini elde ederiz. İşin güzel tarafı bu sefer de $C$ matrisinin
köşegeni altındaki değerler 0. Tıpkı önceki gibi değerleri yerine
koyarak ilerleyebiliriz. Ayşe'nin annesinin yaşını son satırdan
buluruz. Zaten diğerleri de çorap söküğü gibi gelir:

$$
\begin{array}{l}
  \tfrac{1}{2}x_3 = 20 \implies x_3 = 40,\\
  \tfrac{2}{3}x_2 + \tfrac{5}{3}x_3 = 88 \implies x_2 = \tfrac{3}{2}(88 - \tfrac{200}{3}) = 32, \\
  3x_1 + 2x_2 - x_3 = 48 \implies x_1 = \tfrac{1}{3}(48 - 64 + 40) = 8.
\end{array}
$$

Açıkçası $A=BC$ ifadesini sağlayan $B$ ve $C$ matrisleri verilince
problemi çözmek çocuk oyuncağı oldu. İyi de $B$ ve $C$ matrislerini
nasıl elde edeceğiz? Bu sorunun cevabı, bizi bu sayıdaki hesaplamalı
tarife getiriyor. Bir matrisi çarpanlarına ayırmak için önerilmiş
farklı yöntemler var. Aralarından iki tanesini bu yazıda konuşacağız:
LU ve Cholesky ayrışımı. Bu yöntemleri seçmemin iki sebebi var. Bir,
ikisi de oldukça sık kullanılan yöntemler. İki, matris ayrışımı
konusuna bir giriş yapmak için son derece uygunlar. Tarifi anlatmaya
başlamadan önce malzemelerimizi bir listeleyelim.

## Malzemeler

Konuya giriş yaparken matrisleri kullandık bile. Birkaç sembolü daha
tanıtırsak ilerleyen bölümleri yazmamız kolaylaşır. Diyelim elimizde
$m$ satırdan ve $n$ sütundan oluşan bir $A$ matrisi var. Boyutlarını
görmek için $A_{m \times n}$ şeklinde yazabiliriz. Tüm elemanları 0
olan bir matrisi $0_{m \times n}$ ile gösterebiliriz. Eğer bir matrisin
$(i,j)$ elemanından bahsediyorsak, bu $i$. satır ve $j$. sütundaki
değer demektir. Matrisler için çoğunlukla büyük harfler kullanılırken,
elemanları küçük harfle gösterilir. Örneğin $A$ matrisini yazdığımızda
$a_{ij}$ ile de matrisin $(i,j)$ elamanını kastederiz:

$$
A = \left[
  \begin{array}{cccc}
    a_{11}& a_{12} & \dots & a_{1n} \\
    a_{21}& a_{22} & \dots & a_{2n} \\
    \vdots & \vdots & \vdots & \vdots \\
    a_{m1}& a_{m2} & \dots & a_{mn}
  \end{array}
\right].
$$

Matris elemanlarının yerlerini belirlediğimize göre devrik matrisi
tanımlayabiliriz. Yukarda verdiğimiz $A$ matrisinin devriği $A^{\intercal}$ ile
gösterilir. Bu durumda $(i,j)$ elemanı, $(j,i)$ elemanı ile yer
değiştirir. Kısacası $A$ matrisinin satırları sütun, sütunları satır
olunca $A^{\intercal}$ matrisi elde edilir. Yani

$$
A^{\intercal} = \left[
  \begin{array}{cccc}
    a_{11}& a_{21} & \dots & a_{m1} \\
    a_{12}& a_{22} & \dots & a_{m2} \\
    \vdots & \vdots & \vdots & \vdots \\
    a_{1n}& a_{2n} & \dots & a_{mn}
  \end{array}
\right]
$$

olur. Kare matrislerin --adı üstünde-- satır sayısı sütun sayısına
eşittir. Bir kare matrisin simetrik olması devriğinin kendisine eşit
olması demektir. Yeni tanıştığımız sembollerle gösterirsek,
$A_{n\times n}$ matrisi simetrik ise $A = A^{\intercal}$ eşitliğini sağlar.

Köşegen bir matrisin sadece $(1,1)$ ile başlayıp $(m,n)$ ile biten
köşegenindeki sayıları bilmek yeterlidir. Çünkü köşegenleri
haricindeki tüm elemanlar 0 olur:

$$
D = \left[
  \begin{array}{cccc}
    d_{11}& 0 & \dots & 0 \\
    0 & d_{22} & \dots & 0 \\
    \vdots & \vdots & \vdots & \vdots \\
    0 & 0 & \dots & d_{mn}
  \end{array}
\right].
$$

Köşegen matrislerin en meşhuru birim matristir. Birim matris, köşegeni
üzerindeki tüm elemanlar 1 değerini alan kare bir matristir. Boyutunu
da belirtmek için kısaca $I_n$ ile gösterilir.

Sıra geldi matrislerde satır ve sütun işlemlerine. Satırlar ile
başlayalım. Diyelim bir matrisin ilk satırını $\tfrac{2}{3}$ ile
çarpıp ikinci satırdan çıkaracağız. Bir örnek üzerinde hemen
gösterelim:
$$
  \left[\begin{array}{rrr}
        9 & 12 & 24 \\
        6 & 6 & 8 \\
        3 & 4 & 4
       \end{array}\right] \Rightarrow \left[\begin{array}{rrr}
        9 & 12 & 24 \\
        0 & -2 & -8 \\
        3 & 4 & 4
      \end{array}\right].
$$

İşin güzel tarafı bu işlemin aynısını birim matris üzerinde yapıp,
elde ettiğimiz matrisi, asıl matrisimizle soldan çarpabiliriz. Bu da
aynı sonucu verecektir:

$$
\left[\begin{array}{rrr}
        1 & 0 & 0 \\
        -\tfrac{2}{3} & 1 & 0 \\
        0 & 0 & 1
      \end{array}\right]  \left[\begin{array}{rrr}
        9 & 12 & 24 \\
        6 & 6 & 8 \\
        3 & 4 & 4
       \end{array}\right] = \left[\begin{array}{rrr}
        9 & 12 & 24 \\
        0 & -2 & -8 \\
        3 & 4 & 4
      \end{array}\right].
$$

Birim matristen satır işlemleriyle elde ettiğimiz ve soldan çarparken
kullandığımız bu matrislere yalın matrisler denir. Eğer bir matrisin
satırlarını yer değiştirmek istersek, benzer şekilde aynı işlemi birim
matrisin satırları üzerinde yapıp yine soldan asıl matrisimizle
çarpabiliriz. Birim matristen türettiğimiz bu matrislere ise
permütasyon matrisleri denir. Sütun işlemlerine gelirsek, bu işlemler
satır işlemlerine çok benzer. Bu sefer tek yapacağımız bir matrisi
soldan değil de, sağdan yalın matrislerle ya da permütasyon matrisleri
ile çarpmak olur. O durumda hem yalın matrisler, hem de permütasyon
matrisleri birim matrisin sütunları üzerinde ilgili işlemlerin
yapılması ile elde edilirler.

Bir kare matrisin sadece köşegeni ve altındaki değerler 0'dan
farklıysa o matrise alt üçgen matris denir. Böyle bir matrisin devriği
de üst üçgen matris olur. Eğer alt üçgen matrisleri birbirleriyle
çarparsak, ortaya yine alt üçgen bir matris çıkar. Ufak bir örnekte
kolayca deneyebilirsiniz.

Bazı durumlarda bir matrisi alt matrislere bölerek blok matrisler ile
çalışmak isteyebiliriz. Örneğin $A_{m \times n}$ matrisini şu şekilde
4 alt matrise bölebiliriz:

$$
A = \left[
  \begin{array}{ll}
    E_{m_1\times n_1}& F_{m_1\times (n-n_1)}\\
    G_{(m - m_1)\times n_1} & H_{(m - m_1)\times (n\times n_1)}
  \end{array}
\right].
$$

Gelelim bir kare matrisin tersine. Verilen $A_{n \times n}$ matrisi
için 

$$
A A^{-1} = A^{-1} A = I_n
$$

eşitliğini sağlayan bir $A^{-1}$ matrisi bulabiliyorsak, $A$ matrisinin
tersi var demektir. Burada $A^{-1}$ sembolü ile $A$ matrisinin tersi
gösterilir. Bir matrisin --eğer varsa-- tersi tektir. Tersi olmayan
matrislere ise tekil matrisler denir. Bir matris tekil değilse,
çarpanları da tekil değildir. Tekil olmayan bir matrisi herhangi bir
permütasyon matrisi çarparak elde ettiğimiz matrisin de tersi
mevcuttur. Tersi olan bir matrisin, bir sütunu ya da bir satırı
tamamen sıfır olamaz. Yalın alt üçgen matrislerin tersleri de yalın
alt üçgen olurlar. Bu son dört cümlenin kanıtlanması güzel olur. Bir
deneyin bakalım. Blok matrislere dönersek, elimizde

$$
A = \left[
  \begin{array}{ll}
    E_{k\times k} & F_{k \times (n-k)}\\
    0_{(n-k) \times k} & H_{(n - k)\times (n\times k)}
  \end{array}
\right].
$$

şeklinde verilmiş bir üst üçgen matris olsun. Bu matrisin tersinin
olması için gerek ve yeter şart, köşegenleri üzerindeki $E$ ve $H$
matrislerinin tekil olmamalarıdır [1].

Pozitif belirli matrisleri de tanıtırsak malzemelerimiz tamamlanmış
olacak. Elimizde $A_{n \times n}$ olarak verilmiş bir kare matris
olsun. Eğer her $y \neq 0$ vektörü için

$$
y^{\intercal} Ay > 0
$$

eşitsizliği sağlanıyorsa, $A$ matrisi pozitif belirli
demektir. Pozitif belirli bir matrisin köşegenindeki değerlerin tamamı
pozitiftir. Daha da güzeli, pozitif belirli bir matrisin tersi her
zaman vardır. Bir matrisin pozitif belirli olduğunu bulmanın başka
yolları da mevcut. Ancak bu yazının devamı için bu tanım kâfi
gelir. Bir diğer tanımı zaten Bölüm 3'te vereceğiz.


## LU Ayrışımı

İlk konuşacağımız ayrışımın amacı, tersi olduğu bilinen bir
$A_{n \times n}$ matrisini, alt üçgen bir $L_{n \times n}$ matrisi
ile üst üçgen bir $U_{n \times n}$ matrisinin çarpımı şeklinde
yazmak:

$$
A = LU.
$$

Bu sonucu elde etmek için önce $A$ matrisinin köşegeni altındaki
değerleri 0'a eşitlemeye çalışacağız. Bunun için de sırasıyla birkaç
satır işlemi yapacağız. Temel fikri vermek için, bu adımlar sırasında
satırların yerlerinin değişmediğini varsayalım. Kaldı ki yerlerinin
değiştiği durum da benzer şekilde yapılabilir. O konuya bu bölümün
sonundaki teorem ile geleceğim.

Yine bir örnek ile başlamakta fayda var. Diyelim ki matrisimiz

$$
A = \left[\begin{array}{ccc}
        9 & 12 & {24} \\
        \boxed{6} & 6 & {8} \\
        \boxed{3} & \boxed{4} & {4}
      \end{array}\right]
$$

şeklinde verilmiş olsun. Matriste kutu içine aldığım, köşegen
altındaki değerleri 0 yapmaya çalışacağız. Bu sayede üst üçgen $U$
matrisini elde edebiliriz. İlk satırı $\tfrac{2}{3}$ ile çarpıp ikinci
satırdan çıkaralım:

$$
\underset{E_1}{\underbrace{\left[\begin{array}{rrr}
        1 & 0 & 0 \\
        -\tfrac{2}{3} & 1 & 0 \\
        0 & 0 & 1
      \end{array}\right]}} \underset{A}{\underbrace{\left[\begin{array}{rrr}
        9 & 12 & 24 \\
        6 & 6 & 8 \\
        3 & 4 & 4
      \end{array}\right]}}=\left[\begin{array}{rrr}
        9 & 12 & 24 \\
        0 & -2 & -8 \\
        3 & 4 & 4
      \end{array}\right].
$$

Ardından yine ilk satırı alıp, $\tfrac{1}{3}$ ile çarpıp son satırdan çıkararak

$$
\underset{E_2}{\underbrace{\left[\begin{array}{rrr}
        1 & 0 & 0 \\
        0 & 1 & 0 \\
        -\tfrac{1}{3} & 0 & 1
      \end{array}\right]}} \underset{E_1A}{\underbrace{\left[\begin{array}{rrr}
        9 & 12 & 24 \\
        0 & -2 & -8 \\
        3 & 4 & 4
      \end{array}\right]}}=\underset{U}{\underbrace{\left[\begin{array}{rrr}
        9 & 12 & 24 \\
        0 & -2 & -8 \\
        0 & 0 & -4
      \end{array}\right]}}
$$

eşitliğini buluruz. Şansımıza $(3,2)$ konumunda da 0 elde ettik. Eğer
0 olmasaydı, ikinci satırı kullanarak o konumda da 0 çıkmasını
sağlardık. Bu örnek için ikinci satırı 0 ile çarpıp, son satırdan
çıkardık diye düşünebiliriz. Son durumda elimizde şu sistem var:

$$
  E_2E_1A = U.
$$

Dikkat ederseniz, satır değişikliği yapmadığımız için yalın
matrislerimizin tamamı alt üçgen. Bir önceki bölümde bahsettiğimiz
gibi bu matrislerin tersleri de alt üçgen olur. O zaman istediğimiz
sonucu elde ettik demektir:

$$
  A = \underset{L}{\underbrace{E_1^{-1} E_2^{-1}}} U.
$$

Burada $L$ matrisini açıkça yazacak olursak

$$
L =  E_1^{-1} E_2^{-1} = \left[\begin{array}{ccc}
        1 & 0 & 0 \\
        \boxed{\tfrac{2}{3}} & 1 & 0 \\
        \boxed{\tfrac{1}{3}} & \boxed{0} & 1
      \end{array}\right]
$$

eşitliğini elde ederiz. Bir noktaya dikkat etmenizi istiyorum. $A$
matrisinden $U$ matrisine ulaşmaya çalışırken, satırları tam da $L$
matrisinde kutu içersinde gösterdiğim sayılar ile çarpmıştık. Yani $A$
matrisinin ikinci satırında kutu içindeki 6 değerini 0 yapmak için,
$A$ matrisinin ilk satırını $L$ matrisinde aynı konumdaki
$\tfrac{2}{3}$ ile çarptık. Benzer şekilde $A$ matrisindeki 3 değerini
0 yapmak için $L$ matrisinde aynı konumdaki $\tfrac{1}{3}$ çarpanını
kullandık. Yapmamıza gerek kalmadı ama $A$ matrisinin son satırında
kutu içindeki 4 değerini sıfırlamak için $L$ matrisinde aynı
konumdaki 0 değerini kullandık diyebiliriz. Ayrıca üçgen yalın
matrisleri birbirleriyle çarptığımız için $L$ matrisinin köşegeni
sadece 1 değerlerinden oluştu.

Bu gözlem önemli. Çünkü başka herhangi bir matris kullanmadan,
doğrudan $A$ matrisinin üzerine $L$ ve $U$ matrislerinin değerlerini
yazabiliriz. Deneyelim: 

(1) İlk satırı $\tfrac{2}{3}$ ile çarp ve ikinci satırdan çıkar:

$$
  \underset{A}{\underbrace{\left[\begin{array}{rrr}
          9 & 12 & 24 \\
          6 & 6 & 8 \\
          3 & 4 & 4
        \end{array}\right]}} \rightarrow 
         \left[\begin{array}{ccc}
          9 & {12} & {24} \\
          \boxed{\tfrac{2}{3}} & {-2} & {-8} \\
          3 & {4} & {4}
        \end{array}\right].
$$
 
(2) İlk satırı $\tfrac{1}{3}$ ile çarp ve son satırdan çıkar:

 $$
  \left[\begin{array}{ccc}
         9 & {12} & {24} \\
         \boxed{\tfrac{2}{3}} & {-2} & {-8} \\
         3 & {4} & {4}
       \end{array}\right] \rightarrow
        \left[\begin{array}{rrr}
         9 & {12} & {24} \\
         \boxed{\tfrac{2}{3}} & {-2} & {-8} \\
         \boxed{\tfrac{1}{3}} & 0 & {-4}
       \end{array}\right].
 $$
 
(3) Bu adımı yapmaya gerek kalmıyor ama yine de yazalım. İkinci
  satırı $0$ ile çarp ve son satırdan çıkar:
  
 $$
   \left[\begin{array}{ccc}
         9 & {12} & {24} \\
         \tfrac{2}{3} & {-2} & {-8} \\
         \tfrac{1}{3} & {0} & {-4}
       \end{array}\right] \rightarrow
        \left[\begin{array}{rrr}
         9 & {12} & {24} \\
         \boxed{\tfrac{2}{3}} & {-2} & {-8} \\
         \boxed{\tfrac{1}{3}} & \boxed{0} & {-4}
       \end{array}\right].
 $$

En son elde ettiğimiz bu matristeki kutu içine yazılmış sayılar alt
üçgen $L$ matrisinin köşegeni altındaki değerleri, kalanlar da üst
üçgen $U$ matrisinin 0'dan farklı değerlerini veriyor.

Bir örnek verip de işin içinden sıyrılmak olmaz. LU ayrışımının her
zaman çalışacağını matematiksel olarak da göstermeliyiz. Takip eden
teorem tam da bunu yapıyor. Bu teoremin kare olmayan matrisler için
daha genel hali yazılabilir. Zaten ben de öyle bir teoremden
yararlandım [2].

**Teorem 1: ** Elimizde tekil olmayan bir $A_{n \times n}$ matrisi olsun. Bu matrisi,
$P_{n\times n}$ ile gösterilen bir permütasyon matrisinin yardımıyla

$$
P A = L U \hspace{10mm} (2)
$$

  olarak yazabiliriz. Burada $L$ köşegen değerleri bir olan alt üçgen
  bir matris, $U$ ise köşegen değerleri 0'dan farklı bir üst üçgen
  matristir. Ayrıca $P$ matrisi seçildikten sonra (2)
  eşitliğini sağlayacak $L$ ve $U$ matris çifti tektir.

**Kanıt: ** Matrisin boyutları üzerinden tümevarım
yöntemini kullanabiliriz. Başlangıç için $n=1$ alalım. Bu durumda
$L=P=1$ ve $U=A\neq 0$ olur. Şimdi $n > 1$ olduğu durumu inceleyelim.
Elimizdeki $A$ matrisi tekil olmadığı için ilk sütununda 0 olmayan en
az bir eleman vardır. B durumda $\hat{P}$ matrisini öyle bir
seçebiliriz ki $\hat{P} A$ matrisinin $(1,1)$ elemanı 0
olmaz. Kısacası, matrisi alt matrisler kullanarak yazabiliriz:

$$
\hat{P} A =
\left[\begin{array}{cc}
  a & D^{\intercal} \\
  C & B
\end{array}\right].
$$

Burada $a \neq 0$, $C_{(n-1)\times 1}$ ve $D_{(n-1)\times 1}$ ve
$B_{(n-1)\times (n-1)}$ olur. İki çarpan matrisi yardımıyla şu
eşitliği yazabiliriz:

$$
\left[\begin{array}{cc}
  a & D^{\intercal} \\
  C & B
\end{array}\right] = \left[\begin{array}{cc}
  1 & 0 \\
  a^{-1} C & I_{n-1}
\end{array}\right]\left[\begin{array}{cc}
  a & D^{\intercal} \\
  0 & B - a^{-1} CD^{\intercal}
\end{array}\right].
$$

Bu eşitliğin solundaki matris tekil olmadığı için çarpanları da tekil
değildir. Bu durumda eşitliğin sağındaki üst üçgen matrisin alt
köşegenindeki $(n-1) \times (n-1)$ boyutlu matris de tekil olamaz. O
halde, tümevarım hipotezimize göre $\bar{P}$ permütasyon matrisi
kullanılarak elde edilen bir LU ayrışımı mevcuttur. Yani

$$
\bar{P} (B - a^{-1} cd^{\intercal}) = \bar{L}\bar{U}
$$

olur. Bu durumda

$$
P = \left[
  \begin{array}{cc}
    1 & 0_{1 \times (n-1)} \\
    0_{(n-1) \times 1} & \bar{P} 
  \end{array}
\right]\hat{P}
$$

matrisi ile teoreme uygun

$$
L = \left[
\begin{array}{cc}
  1 & 0_{1 \times (n-1)}  \\
  a^{-1} \bar{P} C & \bar{L}
\end{array}
\right]
$$

ve

$$
U = \left[
\begin{array}{cc}
  a & D^{\intercal}  \\
  0_{(n-1) \times 1} & \bar{U}
\end{array}
\right] 
$$

üçgen matrislerini kullanarak (2) eşitliğini elde ederiz.

Kanıtımızın son kısmında (2) ile gösterilen denklemdeki $L$
ve $U$ matris çiftinin tek olduklarını gösterelim. Varsayalım ki $A$
matrisinin $PA = LU$ ve $PA = \bar{L}\bar{U}$ şeklinde iki LU ayrışımı
olsun. Bu durumda $LU = \bar{L}\bar{U}$ olur. $A$ matrisi tekil
olmadığına göre, $L$ ve $U$ çarpanları da tekil değildir. Dolayısıyla

$$
\bar{L}^{-1} L = \bar{U}U^{-1}
$$

eşitliğini yazabiliriz. Bu eşitliğin bir tarafında alt üçgen, diğer
tarafında ise üst üçgen bir matris var. O nedenle eşitlik sadece
çarpımların köşegen matris verdikleri durumda sağlanır.  $L$
matrisinin köşegen değerleri 1 olduğuna göre

$$
\bar{L}^{-1} L = \bar{U}U^{-1} = I_n
$$

eşitliğini ve dolayısıyla $L=\bar{L}$ ve $U=\bar{U}$ sonucunu
göstermiş oluruz. $\Box$



Burada önemli bir noktanın altını çizmeliyim. Teoremde verilen
(2) denklemine göre, $A$ matrisi üzerinde satır yerlerini
değiştirme işlemleri de yapıyoruz. Kabaca söylemek gerekirse bu
işlemlerle, çalışılan matrisin sol üst köşesinde 0'dan farklı bir
değer elde ediyoruz.

## Cholesky Ayrışımı

LU ayrışımı iyi hoş ama ufak bir garipliği var. $L$ matrisinin
köşegen değerleri 1 iken $U$ matrisinin köşegen değerleri 1
değil. Simetrik bir çözüm elde etsek güzel olur. Bunun için $U$
matrisinin her satırını, o satıra karşılık gelen köşegen değere
bölebiliriz. Bölmede kullandığımız sayıları da bir köşegen matrise
taşırız ve yeni $U$ matrisi ile soldan çarparız. Önceki örneğimize
dönersek

$$
\left[\begin{array}{rrr}
        9 & 12 & 24 \\
        0 & -2 & -8 \\
        0 & 0 & -4
      \end{array}\right] = \left[\begin{array}{rrr}
        9 & 0 & 0 \\
        0 & -2 & 0 \\
        0 & 0 & -4
      \end{array}\right]\left[\begin{array}{rrr}
        1 & \tfrac{12}{9} & \tfrac{24}{9} \\
        0 & 1 & \tfrac{-8}{-2} \\
        0 & 0 & 1
      \end{array}\right]
$$

eşitliğini elde ederiz. Bu durumda $A$ matrisi, üç matrisin çarpımı
biçiminde yazılabilir:


$$
A = \left[\begin{array}{rrr}
        1 & 0 & 0 \\
        \tfrac{2}{3} & 1 & 0 \\
        \tfrac{1}{3} & 0 & 1
      \end{array}\right]\left[\begin{array}{rrr}
        9 & 0 & 0 \\
        0 & -2 & 0 \\
        0 & 0 & -4
      \end{array}\right]\left[\begin{array}{rrr}
        1 & \tfrac{12}{9} & \tfrac{24}{9} \\
        0 & 1 & \tfrac{-8}{-2} \\
        0 & 0 & 1
      \end{array}\right].
$$

Genelleyecek olursak, önceki bölümde anlattığımız gibi $A=LU$ şeklinde
bir LU ayrışımı verilmiş olsun. Bu durumda

$$
D = \left[
  \begin{array}{cccc}
    u_{11} & 0      &  \dots & 0 \\
    0      & u_{22} &  \dots & 0  \\
    \vdots & \vdots & \ddots& \vdots \\
    0 & 0  & \dots & u_{nn} 
  \end{array}
\right]
$$

ve

$$
\hat{U}=\left[
  \begin{array}{ccccc}
    1 & \tfrac{u_{12}}{u_{11}}& \tfrac{u_{13}}{u_{11}} & \dots & \tfrac{u_{1n}}{u_{11}} \\
    0 & 1                    & \tfrac{u_{23}}{u_{22}}& \dots&  \tfrac{u_{2n}}{u_{22}}  \\
    0 & 0                  & 1 & \dots& \tfrac{u_{3n}}{u_{33}} \\
    \vdots &  \vdots                  & \vdots & \ddots& \vdots \\
    0 &  0                  & 0 & \dots& 1
  \end{array}
\right]
$$

matrislerini kullanarak

$$
A = LD\hat{U}
$$

eşitliğini yazabiliriz. Eğer $A$ matrisinin simetrik olduğu da
biliniyorsa

$$
A = LDL^{\intercal} \hspace{10mm} (3)
$$

ifadesi elde edilir. Bu son ifadeyi kanıtlamayacağım. Güzel bir ev
ödevi olur ama.

Artık bu bölümün konusu olan ayrışım yöntemine
geçebiliriz. Derdimiz matrislerin köklerini almak. "Kök almak"
ifadesi yeterince açık olmadı. Diyelim elimizde simetrik ve pozitif
belirli bir $A$ matrisi var. Bu matrisi 

$$
A = R^{\intercal} R \hspace{10mm} (4)
$$

şeklinde yazmaya çalışacağız. Bu işleme Cholesky ayrışımı adı
verilir. Dikkate ederseniz $R$ matrisi, $A$ matrisinin
karekökü gibi duruyor. 

Sıradaki teorem bir taşla iki kuş vuruyor. Hem Cholesky ayrışımın
nasıl yapılacağını gösteriyor, hem de bu ayrışımın pozitif belirlilik
testi için biçilmiş kaftan olduğunu söylüyor.

**Teorem 2: ** Simetrik bir $A_{n \times n}$ matrisinin pozitif belirli olması için
gerek ve yeter şart, (4) eşitliğini verecek köşegen
değerleri pozitif bir üst üçgen $R$ matrisinin bulunmasıdır.


**Kanıt: ** Önce gereklilik yönünü gösterelim. $A$
matrisi pozitif belirli olduğu için tekil değildir. Dolayısıyla,
Teorem 1'e göre $A$ matrisinin LU ayrışımı mevcuttur. Ayrıca
$A$ matrisi simetrik de olduğuna göre (3) kullanılarak
$A=LDL^{\intercal}$ şeklinde yazılabilir. Yine $A$ matrisinin pozitif belirli
olması nedeniyle $D$ köşegen matrisinde $d_{ii} > 0$, $i=1, \dots, n$
olur (aksi halde $A$ matrisinin köşegen değerlerinden biri
  negatif olur; alın size bir ev ödevi daha). O zaman

$$
D^{\tfrac{1}{2}}=
\left[
\begin{array}{cccc}
  \sqrt{d_{11}} & 0 & \dots & 0 \\
  0         & \sqrt{d_{22}} & \dots & 0 \\
  \vdots & \vdots & \ddots& \vdots \\
  0 & 0  & \dots & \sqrt{d_{nn}}
\end{array}
\right]
$$

olacak şekilde yeni bir köşegen matris tanımlayabiliriz. Ardından

$$
A = L(D^{\tfrac{1}{2}}D^{\tfrac{1}{2}})L^{\intercal} =
\underset{R^{\intercal}}{\underbrace{(LD^{\tfrac{1}{2}})}}
\underset{R}{\underbrace{(LD^{\tfrac{1}{2}})^{\intercal}}}
$$

ifadesini elde ederiz. Bu eşitlikteki $R$ matrisi, köşegen elemanları
pozitif olan bir üst üçgen matristir. Son kısımda da yeterlilik yönünü
kanıtlayabiliriz. Her $y \neq 0 $ için

$$
y^{\intercal} A y = y^{\intercal} R^{\intercal} R y = (Ry)^{\intercal}(Ry) = \|Ry\|^2 > 0
$$

eşitsizliği elde edilir. Bu da $A$ matrisinin pozitif belirli bir
matris olduğunu gösterir. $\Box$


Dikkat ederseniz, Teorem 2'nin kanıtını Cholesky ayrışımı
yapmak için bir algoritma olarak da düşünebiliriz. Yazması benden,
bilgisayarda denemesi sizden.


## Ve Ötesi

Matris ayrışımı derya gibi konu. Sırf bu iş için yazılmış kitaplar
var [2]. Burada yaptığımız gibi yalnızca LU ve Cholesky ayrışımına
konsantre olsak bile daha söylenecek şeyler var. Ben her iki ayrışım
yöntemi için de kare matrisleri kullandım. Oysa Teorem 1'i,
boyutlara dikkat ederek dikdörtgen matrisler için de pekâlâ
yazabilirdik. Ayrışım konusunu anlatırken hep tersi olan matrisler
kullandım. Ancak bu da şart değil. Örneğin LU ayrışımı yöntemi bir
matrisin tekil olup olmadığını anlamak için de kullanılabilir.

Eğer konu hoşunuza gittiyse QR ayrışıma da bir göz atabilirsiniz
[1]. Bu yazıdan sonra anlamakta zorlanmayacağınızı düşünüyorum. Ancak
önceden dik matrisler konusunda bir alıştırma yapmanız yerinde
olur. Asıl tekil değer ayrışımı diye bir konu var ki pek çok yerde
karşınıza çıkabilir. Onun için biraz hazırlık gerek. İleride vereceğim
tariflerden birini tekil değer ayrışımına ayırmayı planlıyorum.

Yazı boyunca matrislerin çarpanlarından bahsettim. Aranızda matrisler
yerine tam sayıları düşünenler olacaktır. Tam sayıları asal
çarpanlarına tek bir şekilde ayırabiliriz. Ancak matrislerin durumu
öyle değil. Örneğin LU ayrışımı ile elde edilen çarpanlar ile QR
ayrışımı ile elde edilen çarpanlar birbirlerinden farklıdır. 

Bu noktada şunu da söylemeliyim. Matrisleri çarpanlarına ayırmanın tek
sebebi denklem sistemlerini çözmek değil. Bir matrisin özdeğerlerini
bulurken, tersini alırken ya da diğer hesaplamaları yaparken
çarpanları ile çalışmak, kendisi ile çalışmaktan çok daha hızlı
olabilir. 



Bir tarifi daha bitirdik. Bu sefer doğrusal eşitliklerden bahsettik. Bir
dahaki sayıda yine doğrusal denklemler ile uğraşacağız.  Fakat bu

tanımladığı alan üzerinde doğrusal bir fonksiyonun en küçük değerini
bulacağız. Hedefimiz, doğrusal programlamanın temel taşı olan Simpleks

# GNU Octave İşlemleri

Burada uzun uzun anlattığımız ayrışım işlemleri standart lineer cebir paketlerinin tamamında var. Daha [önceki yazıda](http://people.sabanciuniv.edu/sibirbil/MD/MD_SIB_Haziran16.html) da kullandığımız [GNU Octave](https://www.gnu.org/software/octave/) programında bu ayrışım yöntemlerinin nasıl kullanıldığına bir göz atalım. 

Önce bir matris hazırlayalım:

```octave
A = [1 1 1; 2 2 1; 3 2 -1] 
```

Bu yazının başında kullandığımız örnekte karşımıza çıkan matrisin ta kendisi. GNU Octave'da LU ayrışımı için kullanılan fonksiyonun ismi **lu**. Bu fonksiyon bize $L$ ve $U$ matrislerini veriyor.

```octave
[L, U] = lu(A)
```

Bu matrisleri yazının başında biz de çarpanlar olarak kullanmıştık. Dikkat ederseniz $L$ matrisi de üst üçgen veriliyor. Bunun sebebi Teorem 1'deki gibi bir permütasyon matrisi ile çarpıldıktan sonra sonucun verilmesi. Kullanılan permütasyon matrisini de görebiliriz.

```octave
[L, U, P] = lu(A);
P
```

Güzel. İşte bu permütasyon matrisi sayesinde $L$ matrisini istediğimiz şekliyle alt üçgen olarak elde edebiliriz.

```octave
[L, U] = lu(P*A)
```

Son olarak da Cholesky ayrışımına bakalım. Bu ayrışım için kullanılan fonksiyonun ismi de **chol**. Önce pozitif belirli bir $B$ matrisi oluşturalım. 

```octave
B = A'*A
```

Şimdi de bu matrisin "karekökünü" bulabiliriz.

```octave
R = chol(B)
```

Kontrol edelim.

```octave
R'*R
```

Yazının sonunda LU ve Cholesky dışında farklı ayrışım yöntemleri olduğundan bahsetmiştik. Ayrıca bu ayrışım yöntemlerinin bazıları matris tiplerine göre özelleştirilmişlerdir. GNU Octave bu ayrışımların önemli bir kısmını fonksiyonlar olarak sunuyor. [Şu sayfaya](https://www.gnu.org/software/octave/doc/v4.0.0/Matrix-Factorizations.html) bir göz atabilirsiniz.
